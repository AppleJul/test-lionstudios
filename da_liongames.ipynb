### For all statistical tests performed: describe those statistical tests, their assumptions, and why they are the appropriate choice
1. **Z-test for proportions (conversion metrics)**  
- **Purpose:** Compare whether the proportion of users who made a purchase differs between control and test groups.  
- **Assumptions:** Independent users, binary outcome (purchase/no purchase), large enough sample for normal approximation.  
- **Why appropriate:** Conversion is binary, z-test is the standard test for proportions.  

2. **Independent samples t-test (session length)**  
- **Purpose:** Compare average session length per user per day between groups.  
- **Assumptions:** Approximately normal distribution, equal (or Welch-adjusted) variances, independent observations.  
- **Why appropriate:** Session length is a continuous metric, making t-test the classical method to compare means.  

3. **Mann–Whitney U test (non-parametric session length check)**  
- **Purpose:** Validate results without assuming normality.  
- **Assumptions:** Independent samples, data are ordinal/continuous and can be ranked.  
- **Why appropriate:** Session length often has skewed distribution, Mann–Whitney is robust to non-normality.  

4. **Bootstrap confidence intervals (session length differences)**  
- **Purpose:** Estimate distribution of mean differences by resampling.  
- **Assumptions:** Independent observations, no reliance on normality.  
- **Why appropriate:** Provides a robust, distribution-free way to measure uncertainty around mean differences.

### Analysis of Store Icon Clicks

In this part, we evaluate how users interacted with the store icon depending on the test variant.  
The calculations included:  
1. **Average clicks per session** – how often the store icon was opened within a session.  
2. **Average clicks per user** – total clicks aggregated at the user level.  
3. **Click-through rate (CTR)** – share of users who clicked the store icon at least once.  
4. **Statistical test (z-test)** – to compare CTR differences between control and test groups.  
5. **Visualizations** – bar plots were created to compare metrics across variants.  

This step helps quantify user engagement with the new store icon design before analyzing monetization effects.

```python
# Average number of store openings per user and per session. CTR per icon
# how often do users want to return to the store
df = pd.read_csv("lionstudios_data_analyst_take_home.csv")

df['clicked_icon'] = df['icon_shop_opens'] > 0

# avg icon clicks per session
avg_clicks_session = df.groupby('variant')['icon_shop_opens'].mean().reset_index(name='avg_icon_clicks_per_session')
print("Average icon clicks per session:")
print(avg_clicks_session)

# avg icon clicks per user
df_users = df.groupby(['user_id', 'variant'])['icon_shop_opens'].sum().reset_index()
avg_clicks_user = df_users.groupby('variant')['icon_shop_opens'].mean().reset_index(name='avg_icon_clicks_per_user')
print("\nAverage icon clicks per user:")
print(avg_clicks_user)

# CTR users who clicked on icon at least once
df_users['clicked_icon'] = df_users['icon_shop_opens'] > 0
ctr = df_users.groupby('variant')['clicked_icon'].mean().reset_index(name='click_through_rate')
print("\nClick-through rate (icon):")
print(ctr)

# z-test CTR
contingency = df_users.groupby('variant')['clicked_icon'].agg(['sum', 'count'])
success = contingency['sum'].values
nobs = contingency['count'].values
z_stat, p_val = proportions_ztest(success, nobs)
print(f"\nZ-test for icon CTR: z = {z_stat:.4f}, p-value = {p_val:.4f}")

# plots
sns.set(style="whitegrid")

# avg icon clicks per session
plt.figure(figsize=(5, 4))
ax1 = sns.barplot(data=avg_clicks_session, x='variant', y='avg_icon_clicks_per_session', hue='variant', palette='Oranges', legend=False)
plt.title("Avg Icon Clicks per Session")
plt.ylabel("Clicks")
plt.xlabel("")
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.2f}', 
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()

# avg icon clicks per user
plt.figure(figsize=(5, 4))
ax2 = sns.barplot(data=avg_clicks_user, x='variant', y='avg_icon_clicks_per_user', hue='variant', palette='Purples', legend=False)
plt.title("Avg Icon Clicks per User")
plt.ylabel("Clicks")
plt.xlabel("")
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.2f}', 
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()

# icon CTR
plt.figure(figsize=(5, 4))
ax3 = sns.barplot(data=ctr, x='variant', y='click_through_rate', hue='variant', palette='Greens', legend=False)
plt.title("CTR (Clicked Icon ≥1)")
plt.ylabel("Share")
plt.xlabel("")
for p in ax3.patches:
    ax3.annotate(f'{p.get_height():.2%}', 
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

### Shop Icon impact on user behavior  

In this section we prepared and analyzed the A/B test results:  

1. **Data preparation**:  
   - Parsed timestamps for installation and sessions.  
   - Filtered the test period (Jan 1–14, 2023).  
   - Ensured the `variant` field (Control / New Shop Icon) is consistent.  
   - Created user cohorts based on install date.  

2. **Metrics definition**:  
   - Share of users who opened the shop on the **install day**.  
   - Share of users who made a purchase on the **install day**.  

3. **Daily tracking**:  
   - Built daily aggregated tables for both metrics.  
   - Computed per-day z-tests and uplift between Control and Test.  
   - Visualized daily trends (timelines) for shop opens and purchases.  

4. **Pooled analysis**:  
   - Ran z-tests on all days combined to evaluate the overall effect.  

This step was necessary to evaluate whether the new shop icon affected initial engagement (shop opens) and monetization (purchases) compared to the control group.

```python
# the proportion of new users who click on the icon on the first day after installation
# the proportion of new users who make a purchase on the first day after installation
CSV_PATH   = "lionstudios_data_analyst_take_home.csv"
VAR_CONTROL = "Control"
VAR_TEST    = "New Shop Icon"

df = pd.read_csv(CSV_PATH)

def parse_ts(series: pd.Series) -> pd.Series:
    s = pd.to_datetime(series, errors="coerce", utc=True)
    if s.isna().mean() > 0.5:
        best = s
        for unit in ["s", "ms", "us", "ns"]:
            s_try = pd.to_datetime(series, unit=unit, errors="coerce", utc=True)
            if s_try.isna().mean() < best.isna().mean():
                best = s_try
        s = best
    return s

df["install_ts"] = parse_ts(df["install_timestamp"])
df["session_ts"] = parse_ts(df["session_timestamp"])
df = df.dropna(subset=["install_ts", "session_ts"]).copy()

df["install_date"] = df["install_ts"].dt.date
df["session_date"] = df["session_ts"].dt.date

start_date = pd.to_datetime("2023-01-01").date()
end_date   = pd.to_datetime("2023-01-14").date()
df = df[(df["install_date"] >= start_date) & (df["install_date"] <= end_date)].copy()

if "variant" not in df.columns:
    for cand in ["Variant", "VARIANT", "variant_x", "variant_y"]:
        if cand in df.columns:
            df["variant"] = df[cand]
            break
if "variant_x" in df.columns:
    df["variant"] = df.get("variant", df["variant_x"]).fillna(df["variant_x"])
if "variant_y" in df.columns:
    df["variant"] = df.get("variant", df["variant_y"]).fillna(df["variant_y"])

df = df[df["variant"].isin([VAR_CONTROL, VAR_TEST])].copy()

df["icon_shop_opens"] = df.get("icon_shop_opens", 0).fillna(0)
df["icon_purchases"]  = df.get("icon_cash_purchases", 0).fillna(0)

# User cohort
user_cohort = (
    df.groupby("user_id", as_index=False)
      .agg(cohort_install_date=("install_date", "min"),
           variant=("variant", lambda s: s.iloc[0]))
)

install_day_sessions = df.merge(user_cohort, on="user_id", how="inner")
install_day_sessions = install_day_sessions[
    install_day_sessions["session_date"] == install_day_sessions["cohort_install_date"]
].copy()

user_flags = (
    install_day_sessions
    .groupby("user_id", as_index=False)
    .agg(
        shop_open_install=("icon_shop_opens", lambda s: s.sum() > 0),
        purchase_install =("icon_purchases",  lambda s: s.sum() > 0),
    )
)

user_base = (
    user_cohort.merge(user_flags, on="user_id", how="left")
               .fillna({"shop_open_install": False, "purchase_install": False})
)

daily = (
    user_base
    .groupby(["cohort_install_date", "variant"], as_index=False)
    .agg(
        users=("user_id","nunique"),
        users_shopopen=("shop_open_install","sum"),
        users_purchase=("purchase_install","sum"),
    )
    .rename(columns={"cohort_install_date":"install_date"})
    .sort_values("install_date")
)
daily["share_shopopen"] = daily["users_shopopen"] / daily["users"]
daily["share_purchase"] = daily["users_purchase"] / daily["users"]

# z-test
def two_prop_ztest_safe(count1, nobs1, count2, nobs2):
    if (nobs1 is None) or (nobs2 is None) or (nobs1 <= 0) or (nobs2 <= 0):
        return np.nan, np.nan
    counts = np.array([count1, count2], dtype=float)
    nobs   = np.array([nobs1,  nobs2 ], dtype=float)
    if (counts == 0).any() or (counts == nobs).any():
        counts += 0.5
        nobs   += 1.0
    z, p = proportions_ztest(count=counts, nobs=nobs, alternative="two-sided")
    return float(z), float(p)

def build_ab_table(daily_df, metric_prefix):
    ctrl = daily_df[daily_df["variant"]==VAR_CONTROL][
        ["install_date","users",f"users_{metric_prefix}",f"share_{metric_prefix}"]
    ].rename(columns={
        "users":"users_control",
        f"users_{metric_prefix}":"success_control",
        f"share_{metric_prefix}":"share_control"
    })
    test = daily_df[daily_df["variant"]==VAR_TEST][
        ["install_date","users",f"users_{metric_prefix}",f"share_{metric_prefix}"]
    ].rename(columns={
        "users":"users_test",
        f"users_{metric_prefix}":"success_test",
        f"share_{metric_prefix}":"share_test"
    })
    out = pd.merge(ctrl, test, on="install_date", how="outer").sort_values("install_date")

    zs, ps, uplifts = [], [], []
    for _, r in out.iterrows():
        z, p = two_prop_ztest_safe(
            r.get("success_test", np.nan), r.get("users_test", np.nan),
            r.get("success_control", np.nan), r.get("users_control", np.nan)
        )
        zs.append(z); ps.append(p)
        if pd.notna(r.get("share_control")) and r.get("share_control")>0 and pd.notna(r.get("share_test")):
            uplifts.append(100.0*(r["share_test"]-r["share_control"])/r["share_control"])
        else:
            uplifts.append(np.nan)
    out[f"z_{metric_prefix}"] = zs
    out[f"p_{metric_prefix}"] = ps
    out[f"uplift_{metric_prefix}_pct"] = uplifts
    return out.reset_index(drop=True)

table_shop  = build_ab_table(daily, "shopopen")
table_purch = build_ab_table(daily, "purchase")

print("\nA/B per-day table shop open")
print(table_shop)
print("\nA/B per-day table purchase")
print(table_purch)

# plots
shop_pivot = daily.pivot(index="install_date", columns="variant", values="share_shopopen")\
                  .reindex(columns=[VAR_CONTROL, VAR_TEST])
plt.figure(figsize=(12,6))
for var in [VAR_CONTROL, VAR_TEST]:
    if var in shop_pivot.columns:
        y = shop_pivot[var]*100
        plt.plot(shop_pivot.index, y, marker="o", label=var)
        for xi, yi in zip(shop_pivot.index, y):
            if pd.notna(yi):
                plt.text(xi, yi, f"{yi:.1f}%", ha="center", va="bottom", fontsize=9)
plt.title("Shop Open on Install Day (Icon) — Daily Timeline by Variant")
plt.xlabel("Install Day")
plt.ylabel("Share (%)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Purchase %
purch_pivot = daily.pivot(index="install_date", columns="variant", values="share_purchase")\
                   .reindex(columns=[VAR_CONTROL, VAR_TEST])
plt.figure(figsize=(12,6))
for var in [VAR_CONTROL, VAR_TEST]:
    if var in purch_pivot.columns:
        y = purch_pivot[var]*100
        plt.plot(purch_pivot.index, y, marker="o", label=var)
        for xi, yi in zip(purch_pivot.index, y):
            if pd.notna(yi):
                plt.text(xi, yi, f"{yi:.1f}%", ha="center", va="bottom", fontsize=9)
plt.title("Purchase on Install Day (Icon) — Daily Timeline by Variant")
plt.xlabel("Install Day")
plt.ylabel("Share (%)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

def pooled_test(table, label):
    if table[["success_control","users_control","success_test","users_test"]].dropna().empty:
        return
    c_succ = int(table["success_control"].fillna(0).sum())
    c_n    = int(table["users_control"].fillna(0).sum())
    t_succ = int(table["success_test"].fillna(0).sum())
    t_n    = int(table["users_test"].fillna(0).sum())
    z,p = two_prop_ztest_safe(t_succ, t_n, c_succ, c_n)
    ctrl_rate = c_succ / c_n if c_n else np.nan
    test_rate = t_succ / t_n if t_n else np.nan
    uplift = (test_rate - ctrl_rate) / ctrl_rate * 100 if ctrl_rate else np.nan
    print(f"Pooled — {label}: z={z:.4f}, p={p:.4f}, Control={ctrl_rate:.4%}, Test={test_rate:.4%}, Uplift={uplift:.2f}%")

print("\nPooled test shop open")
pooled_test(table_shop, "Shop Open")
print("Pooled test purchase")
pooled_test(table_purch, "Purchase")
```

In this section, the focus is on evaluating **purchases made through the shop icon**.

```python
# number of purchases per user
df = pd.read_csv("lionstudios_data_analyst_take_home.csv")

df['icon_cash_purchases'] = df['icon_cash_purchases'].fillna(0)
df["install_timestamp"] = pd.to_datetime(df["install_timestamp"], errors="coerce", utc=True).dt.tz_localize(None)
df["session_timestamp"] = pd.to_datetime(df["session_timestamp"], errors="coerce", utc=True).dt.tz_localize(None)
df["days_since_install"] = (df["session_timestamp"] - df["install_timestamp"]).dt.days

user_icon_purchases = df.groupby(['user_id', 'variant'])['icon_cash_purchases'].sum().reset_index()
avg_icon_purchases_user = (
    user_icon_purchases.groupby('variant')['icon_cash_purchases']
    .mean()
    .reset_index(name='avg_icon_purchases_per_user')
)

avg_icon_purchases_session = (
    df.groupby('variant')['icon_cash_purchases']
    .mean()
    .reset_index(name='avg_icon_purchases_per_session')
)

user_icon_purchases['made_icon_purchase'] = user_icon_purchases['icon_cash_purchases'] > 0
icon_purchase_rate = (
    user_icon_purchases.groupby('variant')['made_icon_purchase']
    .mean()
    .reset_index(name='conversion_rate_icon')
)

contingency_icon = user_icon_purchases.groupby('variant')['made_icon_purchase'].agg(['sum', 'count'])
success = contingency_icon['sum'].values
nobs = contingency_icon['count'].values
z_stat, p_val = proportions_ztest(success, nobs)

print("=== Overall results ===")
print(avg_icon_purchases_user)
print(avg_icon_purchases_session)
print(icon_purchase_rate)
print(f"Z-test conversion: z={z_stat:.3f}, p={p_val:.4f}\n")

user_day_purchases = (
    df.groupby(['variant', 'user_id', 'days_since_install'])['icon_cash_purchases']
    .sum()
    .reset_index()
)

daily_avg_user_purchases = (
    user_day_purchases.groupby(['variant', 'days_since_install'])['icon_cash_purchases']
    .mean()
    .reset_index(name="avg_icon_purchases_per_user")
)

user_day_purchases['made_icon_purchase'] = user_day_purchases['icon_cash_purchases'] > 0
daily_conversion_purchases = (
    user_day_purchases.groupby(['variant', 'days_since_install'])['made_icon_purchase']
    .mean()
    .reset_index(name="conversion_rate_icon")
)

# charts
sns.set(style="whitegrid")

plt.figure(figsize=(5, 4))
ax1 = sns.barplot(data=avg_icon_purchases_user, x='variant', y='avg_icon_purchases_per_user', color='orange')
plt.title("Avg Icon Purchases per User (Overall)")
plt.ylabel("Purchases")
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.2f}', (p.get_x()+p.get_width()/2., p.get_height()), 
                 ha='center', va='bottom', fontsize=10)
plt.show()

plt.figure(figsize=(5, 4))
ax2 = sns.barplot(data=avg_icon_purchases_session, x='variant', y='avg_icon_purchases_per_session', color='purple')
plt.title("Avg Icon Purchases per Session (Overall)")
plt.ylabel("Purchases")
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.2f}', (p.get_x()+p.get_width()/2., p.get_height()), 
                 ha='center', va='bottom', fontsize=10)
plt.show()

plt.figure(figsize=(5, 4))
ax3 = sns.barplot(data=icon_purchase_rate, x='variant', y='conversion_rate_icon', color='green')
plt.title("Conversion to Icon Purchases (Overall)")
plt.ylabel("Share")
for p in ax3.patches:
    ax3.annotate(f'{p.get_height():.2%}', (p.get_x()+p.get_width()/2., p.get_height()), 
                 ha='center', va='bottom', fontsize=10)
plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(data=daily_avg_user_purchases, x="days_since_install", y="avg_icon_purchases_per_user", hue="variant", marker="o")
for variant in daily_avg_user_purchases["variant"].unique():
    subset = daily_avg_user_purchases[daily_avg_user_purchases["variant"] == variant]
    for x, y in zip(subset["days_since_install"], subset["avg_icon_purchases_per_user"]):
        plt.text(x, y+0.01, f"{y:.2f}", ha="center", fontsize=8)
plt.title("Avg Icon Purchases per User by Day")
plt.ylabel("Purchases")
plt.xlabel("Days Since Install")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(data=daily_conversion_purchases, x="days_since_install", y="conversion_rate_icon", hue="variant", marker="o")
for variant in daily_conversion_purchases["variant"].unique():
    subset = daily_conversion_purchases[daily_conversion_purchases["variant"] == variant]
    for x, y in zip(subset["days_since_install"], subset["conversion_rate_icon"]):
        plt.text(x, y+0.002, f"{y:.2%}", ha="center", fontsize=8)
plt.title("Conversion to Icon Purchases by Day")
plt.ylabel("Share")
plt.xlabel("Days Since Install")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

def run_pooled_test(df, metric_col, event_name="Metric"):
    user_data = df.groupby(['user_id', 'variant'])[metric_col].sum().reset_index()
    user_data['converted'] = user_data[metric_col] > 0
    
    contingency = user_data.groupby('variant')['converted'].agg(['sum', 'count'])
    
    control_success, control_total = contingency.loc['Control']
    test_success, test_total = contingency.loc['New Shop Icon']
    
    control_rate = control_success / control_total
    test_rate = test_success / test_total
    uplift = (test_rate / control_rate - 1) * 100
    
    success = [control_success, test_success]
    nobs = [control_total, test_total]
    z_stat, p_val = proportions_ztest(success, nobs)
    
    print(f"=== {event_name} ===")
    print(f"Control: {control_success}/{control_total} = {control_rate:.2%}")
    print(f"New Shop Icon: {test_success}/{test_total} = {test_rate:.2%}")
    print(f"Uplift: {uplift:.2f}%")
    print(f"Z-test: z={z_stat:.4f}, p={p_val:.4f}\n")

run_pooled_test(df, "icon_shop_opens", event_name="Icon Shop Opens")
run_pooled_test(df, "icon_cash_purchases", event_name="Icon Purchases")
```

## Retention Analysis (Control vs New Shop Icon)

We calculate **daily user retention** to compare engagement between groups.

**Steps:**
1. Convert timestamps and align timezones.  
2. Compute `days_since_install` = session day – install day.  
3. Define new users (Day 0).  
4. For each day and variant:
   - `active_users` = users with a session,  
   - `new_users` = all Day 0 users,  
   - `retention_rate` = active_users / new_users * 100.  
5. Plot retention curves with annotated values.

**Goal:**  
See if **New Shop Icon** improves retention vs **Control**, especially in early days.

```python
# retention
df = pd.read_csv("lionstudios_data_analyst_take_home.csv")

df["install_timestamp"] = pd.to_datetime(df["install_timestamp"], errors="coerce", utc=True).dt.tz_localize(None)
df["session_timestamp"] = pd.to_datetime(df["session_timestamp"], errors="coerce", utc=True).dt.tz_localize(None)

df["days_since_install"] = (df["session_timestamp"] - df["install_timestamp"]).dt.days
df = df[df["days_since_install"] >= 0]

# retention
user_day = df.groupby(["variant", "user_id", "days_since_install"]).size().reset_index(name="sessions")

new_users = user_day[user_day["days_since_install"] == 0][["variant", "user_id"]]

retention = (
    user_day.groupby(["variant", "days_since_install"])["user_id"]
    .nunique()
    .reset_index(name="active_users")
)

retention = retention.merge(
    new_users.groupby("variant")["user_id"].nunique().reset_index(name="new_users"),
    on="variant"
)

retention["retention_rate"] = retention["active_users"] / retention["new_users"] * 100

# charts
plt.figure(figsize=(12,6))
sns.lineplot(data=retention, x="days_since_install", y="retention_rate", hue="variant", marker="o")

for variant in retention["variant"].unique():
    subset = retention[retention["variant"] == variant]
    for x, y in zip(subset["days_since_install"], subset["retention_rate"]):
        plt.text(x, y+0.5, f"{y:.1f}%", ha="center", fontsize=8)

plt.title("Daily User Retention by Variant")
plt.xlabel("Days Since Install")
plt.ylabel("Retention Rate (%)")
plt.xticks(range(0, int(retention["days_since_install"].max()) + 1))
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="Variant")
plt.show()
```

### Analysis: Average Session Length per User per Day
**What we did:**  
- Calculated the average session length per user per day for each variant.  
- Visualized daily trends with a line chart.  
- Compared Control vs. New Shop Icon using statistical tests.  
**Statistical tests applied:**  
- T-test checks mean differences assuming normality.  
- Mann–Whitney U non-parametric alternative, robust to non-normal distributions.  
- Bootstrap confidence interval estimated the average difference in minutes with 95% CI.  
Why: Determine whether users in the New Shop Icon group spend significantly more (or less) time per session compared to Control

```python
# session length
df = pd.read_csv("lionstudios_data_analyst_take_home.csv")

df["session_timestamp"] = pd.to_datetime(df["session_timestamp"], errors="coerce", utc=True).dt.tz_localize(None)
df["session_date"] = df["session_timestamp"].dt.date

user_day = (
    df.groupby(["variant", "user_id", "session_date"])["session_length"]
    .mean()
    .reset_index(name="avg_session_length_user_day")
)

daily_avg = (
    user_day.groupby(["variant", "session_date"])["avg_session_length_user_day"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(14,6))
sns.lineplot(data=daily_avg, x="session_date", y="avg_session_length_user_day", hue="variant", marker="o")

for variant in daily_avg["variant"].unique():
    subset = daily_avg[daily_avg["variant"] == variant]
    for x, y in zip(subset["session_date"], subset["avg_session_length_user_day"]):
        plt.text(x, y+0.3, f"{y:.1f}", ha="center", fontsize=8)

plt.title("Average Session Length per User per Day")
plt.ylabel("Avg Session Length (minutes)")
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="Variant")
plt.show()

control = user_day[user_day["variant"] == "Control"]["avg_session_length_user_day"]
new_icon = user_day[user_day["variant"] == "New Shop Icon"]["avg_session_length_user_day"]

# t-test
t_stat, p_val_t = ttest_ind(control, new_icon, equal_var=False)

# Mann-Whitney U
u_stat, p_val_u = mannwhitneyu(control, new_icon, alternative="two-sided")

print("Statistical tests for avg session length per user per day:")
print(f"T-test: t={t_stat:.3f}, p={p_val_t:.4f}")
print(f"Mann-Whitney: U={u_stat}, p={p_val_u:.4f}")

def bootstrap_ci(data1, data2, n_boot=10000, ci=95):
    diffs = []
    for _ in range(n_boot):
        sample1 = np.random.choice(data1, size=len(data1), replace=True)
        sample2 = np.random.choice(data2, size=len(data2), replace=True)
        diffs.append(sample2.mean() - sample1.mean())
    lower = np.percentile(diffs, (100-ci)/2)
    upper = np.percentile(diffs, 100-(100-ci)/2)
    return np.mean(diffs), (lower, upper)

mean_diff, ci_diff = bootstrap_ci(control.values, new_icon.values)
print(f"\nBootstrap mean difference (New Icon - Control): {mean_diff:.3f} minutes")
print(f"{95}% CI: [{ci_diff[0]:.3f}, {ci_diff[1]:.3f}] minutes")
```
